### An Agent can Store past memory

```python
from dotenv import load_dotenv

from langchain_tavily import TavilySearch
from langchain_groq import ChatGroq
from langchain.agents import create_agent
#? agent ke ander memory save krne ke liye => ye InMemorySaver use hota hai 
from langgraph.checkpoint.memory import InMemorySaver

load_dotenv()

llm = ChatGroq(model="openai/gpt-oss-20b")


tavily_search_tool = TavilySearch(
    max_results = 3,
    topic = "general"
)

agent = create_agent(
    model= llm,
    tools=[tavily_search_tool],
    system_prompt="You are a agent and can search any query on internet",
    #**  ye checkpointer lagana padega !! 
    checkpointer=InMemorySaver()
)

#*** NOW AGENT IS READY => just ab mujhe agent ko invoke krna hai with any user question !! 

#user_query = "who is biswa bijaya samal?"
user_query = "who won ipl 2026 in one sentence?"


# TODO => ab yahan pr => thread_config = {"configurable": {"thread_id": "1"}} ye krna padega isse kya hoga ki 
# TODO => har thread_id ek memeory rakhega store apne mein => jaise he we change thread id hum dusre memory mein ajyengein !

res = agent.invoke(
    {"messages":[{"role":"user" , "content":user_query}]},
    {"configurable":{"thread_id": "first"}}
    )

res["messages"][-1].content
```
