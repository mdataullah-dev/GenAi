### RAG => Retreival Augmented Generation


```python
# rag is a technique

#? llm + external knowledge bases  ( provide external data)

#? example => giving a llm a book => then llm will find ans or accurate ans from this book rather than using training data

#***  RAG PIPELINE (this is 1 time process)
"""
STEP1. document/data loading

STEP2. Split the data/docs in chunks

STEP3. Vector Embeddings

STEP4. Vector DB => it stores vectores embeddings
"""
#! LLM works on numerical data not on texts!!
#? jo v data -> chunks mein divide ho gya hai -> usi ka numerical representation hai vector embeddings


#*** Now what will happen when User ask Question

"""
STEP1. USER ASK QUESTION
#? us ques ko first VECTOR DB ke pass bhejengein and usne se similar record fetch kr ke layengein
#? then based on user question -> we(hum) aone VECTOR DB is most matchable or suitable record find krengein jo user ke question ke a/c ho ya question se milta julta ho
#? means vector db se vo data hum fetch krengein jo user ke question ke similar ho 
#* and this process is called as similarity search ( where we apply similarity search algoritm)
#? then jo vector DB hum log ko dega we will store it in any variable

STEP2. STORING RESPONSE FROM VECTOR DB INTO SIMILAR DATA VARIABLE

STEP3. NOW A LLM WILL USE THIS SIMILAR DATA (As context) + USERS QUESTION

STEP4. THEN LLM WILL GENERATE FINAL ANSWER AND GIVE IT TO USER
"""

#!! THIS IS WHOLE RAG PIPELINE or RAG BASED SYSTEM
```

```python
#1 . STEP 1 : DATA LOADING 

#? Using langchain framework we will do this
#***  inside langchain docs -> integrations -> in integration by components -> data loaders

#? load() => loads all documen t at once 
#? lazy_load() => streams documents lazily , useful for large datasets
```
