```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-120b")
```

Discussion of thr problem :

```python
text = "My name is Md Ataullah, "\
"email is work@gmail.com, and my age is 22"   #? backslash "\" means continues lines

text1 = ("My name is Md Ataullah "
"email is work@gmail.com, and my age is 22")  #* when inside () => no need of "\"

print(text)
print(text1)
```

```python
res = llm.invoke(f'Give me only name , email , age from this text: {text}')
print(res.content)
print(type(res.content))  #? its string type of content ==> object | dict type nhi hai 
```

Solution : structured Output

```python
from pydantic import BaseModel, Field

class ResponseStructure(BaseModel):
    # name: str = Field(description= "Name of the user")    #? doing feild validation
    # email: str = Field(description= "Email of the user") 
    # age: int = Field(description= "Age of the user") 
    name: str = Field(description= "Complete Name")    #? doing feild validation
    email: str = Field(description= "Email Address") 
    age: int = Field(description= "Age") 
    
'''
LLM model ko har feild ka datatype bata diya ==> ab har field ka description v dena hoga 
LLM ko ==> ki kis feild meim kya dalna hai ==> uske liye we use Field  
'''

structured_llm = llm.with_structured_output(ResponseStructure)

res = structured_llm.invoke(f'Give me only name , email , age from this text: {text}')

print(type(res))
res
# res.age
# res.email
```
