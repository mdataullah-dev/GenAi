### Stream LLM Responses in Real-Time like Chat-gpt! 

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-120b", streaming=True)

question = "Can you explain me gen-ai in short"
response = llm.stream(question)

for chunks in response:
    print(chunks.content, end="")
```
