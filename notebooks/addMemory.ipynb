Add Memory to AI

```python
from dotenv import load_dotenv
load_dotenv()

```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-120b")
```

```python
res = llm.invoke("Hello, My name is MD ATAULLAH MASOOD. I am a software engineer. I have 1 year of experience in software development")
print(res.content)
                 
```

```python
res = llm.invoke("What is my name?")
print(res.content)

##* this is memory test ! 
```

```python
## how to add memory !!

prompts = [
    {"role":"user" , "content":"hello, My name is Md Ataullah Masood"}
]

res = llm.invoke(prompts)
print(res.content)
```

```python
## Hello, Md Ataullah Masood! 👋 Nice to meet you. How can I assist you today?

#*** passing this answer of ai agian as prompt 

prompts = [
    {"role":"user" , "content":"hello, My name is Md Ataullah Masood"},
    {"role":"ai" , "content":"Hello, Md Ataullah Masood! 👋 Nice to meet you. How can I assist you today?"},
    {"role":"user" , "content":"What is my name?"}
]

res = llm.invoke(prompts)
print(res.content)
```

## Manage History - passing previous msg of ai dynamically


```python
history = [] 

while True:
    
    query = input("User: ")
    if query.lower() in ["exit" , "stop"]:
        print("OK!")
        break
    
    history.append({"role":"user" , "content": query})     ###? for example ==> query will be "Who is pm of india?"  ==> then it(dict) will append in history list
    #print("User: ", query)
    print(f'USER : {query}')
    
    res = llm.invoke(history)
    
    history.append({"role":"ai" , "content": res.content})
    print(f'AI : {res.content}')
    
```
