Dynamic Prompts

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

from langchain_core.output_parsers import StrOutputParser


'''
||  Prompts Templates === Dynamic Prompts ||

#*  ChatPromptTemplate ==> gives permission to make dynamic prompts
'''

llm = ChatGroq(model="openai/gpt-oss-120b")

content = StrOutputParser()

prompts = ChatPromptTemplate([
    {"role":"system", "content":"You are a translator and translate input to {language}"},
    {"role":"user", "content":"{query}"},
])

prompts

# res = llm.invoke(prompts)
# print(res.content)
```

```python
final_prompts = prompts.format_messages(language = "hindi" , query = "Why this ai world exits!!")
final_prompts
```

```python
res = llm.invoke(final_prompts)
print(res.content)
```

```python
final_prompts = prompts.format_messages(language = "arabic" , query = "Why this ai world exits!!")
res = llm.invoke(final_prompts)
print(res.content)
```

```python
'''
||Chains === Runnable ||

#* runnable ==> jo function ya jo data ko hum invoke kr skte hain vo runnable hai
'''

# prompts.invoke({"language":"Hinglish" , "query":"I will learn nodejs"})   #=> passed as chain inside llm.invoke

res = llm.invoke(prompts.invoke({"language":"Hinglish" , "query":"I will learn nodejs"}).messages)    ## here , .messages will give me the msg that prompts.invoke give

#/ messages=[SystemMessage(content='You are a translator and translate input to Hinglish', additional_kwargs={}, response_metadata={}), HumanMessage(content='I will learn nodejs', additional_kwargs={}, response_metadata={})]
res.content

#! This is chaining 
```

```python
###   Self chain function

def transform_case(result:str):
    return result.upper()
```

Correct way to chaining

```python
chains = prompts | llm | content | transform_case      #*   chaining done => using pipe ==> no need of above code for doing chaining 

#!  here prompts --> will go automatically to llm

print(chains)

print("===   ===   ===")

res = chains.invoke({"language":"Hindi" , "query":"I will learn nodejs"})
print(res)
print("===   ===   ===")
#print(res.content)  #? no need to type .content => use strparser
print("===   ===   ===")
```
