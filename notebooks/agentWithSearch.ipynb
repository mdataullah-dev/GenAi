```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent

llm = ChatGroq(model="openai/gpt-oss-20b")
```

```python
tavily_search_tool = TavilySearch(
    max_results = 3,
    topic = "general"
)

agent = create_agent(
    model= llm,
    tools=[tavily_search_tool],
    system_prompt="You are a agent and can search any query on internet"
)

#*** NOW AGENT IS READY => just ab mujhe agent ko invoke krna hai with any user question !! 

user_query = "who is biswa bijaya samal?"

res = agent.invoke({"messages":[{"role":"user" , "content":user_query}]})

len(res["messages"])

res["messages"]
```
