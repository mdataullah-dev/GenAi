```python
from langchain.tools import tool
#from langchain_core.tools import tool
```

```python
@tool
def add_num(a:int, b:int):
    """
    it will return the sum of two numbers
    Args:
        a : Number 1
        b : Number 2
    """
    return a+b

#* now how to use tools ? 
#? => for using tool => we use invoke() 
```

```python
@tool
def mul_num(a:int, b:int):
    """
    it will return the multiplication of two numbers
    Args:
        a : Number 1
        b : Number 2
    """
    return a*b
```

```python
add_num.invoke({"a":4 , "b":5})
mul_num.invoke({"a":4 , "b":5})
    
#? now connect -> tools with LLM => agents 
```

```python
from langchain.agents import create_agent
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-120b")
```

```python
import langchain.agents as agents
print(dir(agents))
```

```python
agent = create_agent(
    model = llm,
    tools=[add_num , mul_num],
    system_prompt="You are a math teacher and you have to always use tools for calculation"
)
```

```python
response = agent.invoke({"messages":[{"role":"user" , "content": "What is 3 + 9"}]})
```

```python
response
```

```python
for res in response["messages"]:
    print(res)
    print("\n")
```
